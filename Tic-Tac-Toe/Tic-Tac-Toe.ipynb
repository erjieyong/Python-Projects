```python
def display(positions):
    row0 = '     1   2   3\n'
    row1 = '  |   |   |   |\n'
    row2 = f'A | {positions["A1"]} | {positions["A2"]} | {positions["A3"]} |\n'
    row3 = '  |___|___|___|\n'
    row4 = '  |   |   |   |\n'
    row5 = f'B | {positions["B1"]} | {positions["B2"]} | {positions["B3"]} |\n'
    row6 = '  |___|___|___|\n'
    row7 = '  |   |   |   |\n'
    row8 = f'C | {positions["C1"]} | {positions["C2"]} | {positions["C3"]} |\n'
    row9 = '  |   |   |   |\n'
    print(row0,row1,row2,row3,row4,row5,row6,row7,row8,row9)
    
```

```python
def initialise():
    positions = {'A1':' ','A2':' ','A3':' ','B1':' ','B2':' ','B3':' ','C1':' ','C2':' ','C3':' '}
    
    player1 = {
        'name':'',
        'sign':'',
        'position':[]
    }
    player2 = {
        'name':'',
        'sign':'',
        'position':[]
    }
    
    while player1['name']=='':
        player1['name'] = input("Enter Player 1's name:")
    while player1['sign']=='' or len(player1['sign']) != 1:
        player1['sign'] = input("Player 1 choose your sign (1 char only):").upper()
    while player2['name']=='':
        player2['name'] = input("Enter Player 2's name:")
    while player2['sign']=='' or len(player2['sign']) != 1 or player2['sign'] == player1['sign']:
        player2['sign'] = input("Player 2 choose your sign (1 char only):").upper()
    
    return positions, player1, player2
```

```python
def game_won(player):
    # check for 3 of the same columns or rows
    column_row_match = False
    character_list = []
    for position in player['position']:
        character_list.extend(position)
    results = {'A':0,'B':0,'C':0,'1':0,'2':0,'3':0}
    
    for char in character_list:
        results[char] += 1
    
    column_row_match = 3 in results.values()

    #check for diagonal lines
    diagonal_match = False
    match1 = ['A1', 'B2', 'C3']
    match2 = ['A3', 'B2', 'C1']
    for position in player['position']:
        if position in match1:
            match1.remove(position)
        if position in match2:
            match2.remove(position)
    if len(match1) == 0 or len(match2) == 0:
        diagonal_match = True
    
    return column_row_match or diagonal_match
```

```python
def play_game(start):
    positions, player1, player2 = start
    current_player = player1
    current_position_chosen = ''
    display(positions)
    while True:
        current_position_chosen = input(f'{current_player["name"]} choose your position:')
        while current_position_chosen not in positions or current_position_chosen in player1['position'] or current_position_chosen in player2['position']:
            current_position_chosen = input(f'{current_player["name"]} choose your position:')
        current_player['position'].append(current_position_chosen)
        positions[current_position_chosen] = current_player['sign']
        display(positions)
        if game_won(current_player):
            print(f"{current_player['name']} won!!!")
            if input('Do you want to play again? (Y or N)') == 'Y':
                play_game(initialise())
            else:
                break
        elif ' ' not in positions.values():
            print("Nobody wins.")
            if input('Do you want to play again? (Y or N)') == 'Y':
                play_game(initialise())
            else:
                break
        elif current_player == player1:
            current_player = player2
        else:
            current_player = player1
```
